Merging and joining of dataframes

```python
df1 = pd.DataFrame({'a':[2,4,23,4,6,2],'b':['d','f','ws','e','w','y']})
df1
```

```python
df3 = pd.DataFrame({'b':['d','f',230,40,60,200,300],'d':['dasd','fg','wsgh','eaf','aew','hjy','fg']})
df3
```

```python
pd.merge(df1,df3)  #default is inner  - only common ones
```

```python
pd.merge(df1,df3, how = 'left') #Everything from left df is taken and only common ones from right, rest kept as NaN
```

```python
pd.merge(df1,df3, how = 'right')
```

```python
pd.merge(df1,df3, how = 'outer') #Everything from both sides
```

```python
df2 = pd.DataFrame({'e':[20,40,230,40,60,200],'f':['dasd','fg','f','d','e','hjy']})
df2
```
