```python
# we need to separate independent variables and a dependent variable
X=df.drop(columns="tip",axis=1)
y=df["tip"]
```

```python
# split the dataset into train and test
 # 80% train and rest 20 % for testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0) # random state is for reprodcibility
```

```python
x_test # feature scaling, feature selection, cross validation, hyper parameter tuning
```

```python
y_test # 20
```

```python
# Model Training

from sklearn.linear_model import LinearRegression
lin_r=LinearRegression()
lin_r.fit(x_train,y_train) # 80
```

```python
# evaluation using test dataset
#y_test # 20 # x_test
y_pred=lin_r.predict(x_test)
```

```python
y_pred
```
